# Jean pockets

Are women jean pockets ridiculously small? Let's look at the data!

The file `jean-pocket-measurements.csv` shows pocket measurements for 20 popular brands.
Four pairs of jeans from each brand were measured: men's and women's skinny and straight styles. All jeans were designated a 32-inch waistband. 

The following measurements are included:
- `height_front`: maximum front pocket height (cm)
- `height_back`: maximum back pocket height (cm)
- `width_front`: maximum front pocket width (cm)
- `width_back`: maximum back pocket width (cm)

```python
import pandas as pd
df = pd.read_csv('jean-pocket-measurements.csv')
df.head()
```

### Project Ideas

- What is the average difference in pocket `height_front` between women's and men's jeans?

- Is there a significant difference in pocket `height_front` between skinny and straight styles within the same gender?

- How do back pocket sizes compare between women's and men's jeans?

- Measure your phone's height. What percentage of women's and men's jeans can comfortably fit your phone in the pockets?



### Solutions

To determine the average difference in pocket **height_from** between women's and men's jeans I am going to apply a **groupby()** and them calculate the average. Functions used:
- gropby(): To group by gender
- mean(): To calculate the average
- reset_index(): To convert a serie into dataframe
- query(): To search by gender



```python
average_df=df.groupby('gender')['height_front'].mean().reset_index()

men_height = average_df.query('gender == "men"')['height_front'].values[0]
women_height = average_df.query('gender == "women"')['height_front'].values[0]
result = abs(men_height - women_height).round(2)

print(result)
```

The average different is 8.8 cm.

Now, I want to calculate the differente between skinny and straight styles, divided by gender

```python
grouped_df = df.groupby(['gender', 'style'])['height_front'].mean().reset_index()

height_front_col_M = grouped_df.query('gender == "men"')['height_front']

height_front_col_W = grouped_df.query('gender == "women"')['height_front']

result_M = abs(height_front_col_M.values[0] - height_front_col_M.values[1])

result_W = abs(height_front_col_W.values[0] - height_front_col_W.values[1])

print("Pocket different between skinny and straight men pants: %.2f cm" %result_M)
print("Pocket different between skinny and straight women pants: %.2f cm" %result_W)

grouped_df
```

From previous calculation we can conclude that there isn't a big different in pockets between skinny and straight pants in both genders.

#### Back pockets

Find out the different in back pockets size between men and women

```python
#area
df['area']=df.eval('height_back*width_back')

area_df = df.groupby('gender')['area'].mean().reset_index()

result = abs(area_df['area'].values[0]-area_df['area'].values[1])

print("Back pockets differents: %.2f cm" %result)
```

The different between back pockets is big.
